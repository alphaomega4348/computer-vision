```python
import os
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

list_name = os.listdir("/content/images")
print(list_name)

for name in list_name:
  # Check if the item in the list is a file and not a directory (like '.ipynb_checkpoints')
  # and if it's an image file by checking the file extension.
  if os.path.isfile(os.path.join("/content/images", name)) and name.lower().endswith(('.png', '.jpg', '.jpeg')):
    img = cv2.imread(f"/content/images/{name}")
    if img is not None:  # Check if image was loaded successfully
      # Display the original image
     # cv2_imshow(img)

      # Resize the image
      resized_img = cv2.resize(img, (300, 300))


      cv2_imshow(resized_img)

      cv2.waitKey(0)
      cv2.destroyAllWindows()
    else:
      print(f"Failed to load image: {name}")
```
